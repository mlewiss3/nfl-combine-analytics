```python
import nflreadpy as nfl

combine_data = nfl.load_combine()
combine_df = combine_data.to_pandas()

combine_df.head()
```

```python
combine_df['pos'].unique()
```

```python
my_positions = ['CB', 'S', 'WR']
filtered_combine = combine_df[combine_df['pos'].isin(my_positions)]

filtered_combine.shape
```

```python
player_stats = nfl.load_player_stats(seasons=True)
stats_df = player_stats.to_pandas()

stats_df.head()
```

```python
player_stats_offense = nfl.load_player_stats(seasons=True, stat_type='offense')
stats_offense_df = player_stats_offense.to_pandas()

stats_offense_df.columns
```

```python
help(nfl.load_player_stats)
```

```python
player_stats_all = nfl.load_player_stats(seasons=True, summary_level='reg')
stats_all_df = player_stats_all.to_pandas()

stats_all_df.columns
```

```python
for col in stats_all_df.columns:
    print(col)
```

```python
my_positions = ['CB', 'S', 'WR']

cols_to_keep = [
    'player_id', 'player_name', 'position', 'season', 'recent_team', 'games',
    'receptions', 'targets', 'receiving_yards', 'receiving_tds',
    'def_interceptions', 'def_pass_defended', 'def_tackles_solo',
    'def_sacks', 'def_interception_yards'
]

filtered_stats = stats_all_df[stats_all_df['position'].isin(my_positions)][cols_to_keep]

filtered_stats.shape
```

```python
filtered_combine.to_csv('combine_data_cb_s_wr.csv', index=False)
filtered_stats.to_csv('player_season_stats.csv', index=False)

print("Both files saved successfully")
```
